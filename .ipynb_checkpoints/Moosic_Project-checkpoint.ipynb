## Silhouette

```python
# ALL FOUR SCALERS TO SILHOUETTE
# songs_scaled_minmax
# songs_scaled_maxAbs
# songs_scaled_rbs
# songs_scaled_quantile
scalers = [songs_scaled_minmax,songs_scaled_maxAbs,songs_scaled_rbs,songs_scaled_quantile]
```

```python
def get_scalers(ls_df):
    ls_df = scalers
    return ls_df
```

```python
# def get_title()
```

```python
def silhouette_plot():    
    scalers_list = get_scalers(scalers)
    title_list = ["MinMax Scaler","MaxAbs Scaler","Robust Scaler","Quantile Scaler"]
#     max_k = [25,50,75,100]
    max_k = [50,100]
    min_k = 10
#     max_k = [5,10]
    for k,item in enumerate(scalers_list):
        for i in max_k:
            sil_score = []
            for j in range(min_k, i): 
                labels = KMeans(n_clusters=j).fit(scalers_list[k]).labels_
                sil_score.append(silhouette_score(scalers_list[k], labels))

            plt.title(f'Silhouette score evolution from {min_k} cluster to {i} clusters with {title_list[k]}')
            sns.lineplot(x=range(min_k, i), y=sil_score)
            sns.scatterplot(x=range(min_k, i), y=sil_score)
            plt.show()
```

```python
silhouette_plot()
```

```python
def single_min_max_sil(x,y):
    sil_score = []
    if x<y :
        for k in range(x, y): 
            labels = KMeans(n_clusters=k).fit(songs_scaled_minmax).labels_
            sil_score.append(silhouette_score(songs_scaled_minmax, labels))
        plt.title(f'Silhouette score evolution from {x} cluster to {y} clusters')
        sns.lineplot(x=range(x, y), y=sil_score)
        sns.scatterplot(x=range(x, y), y=sil_score)
        plt.show()
    if x>y:
        for k in range(y, x): 
            labels = KMeans(n_clusters=k).fit(songs_scaled_minmax).labels_
            sil_score.append(silhouette_score(songs_scaled_minmax, labels))
        plt.title(f'Silhouette score evolution from {y} cluster to {x} clusters')
        sns.lineplot(y=range(y, x), x=sil_score)
        sns.scatterplot(y=range(y, x), x=sil_score)
        plt.show()
```

```python
def single_min_max_sil(x,y):
    sil_score = []
    if x<y :
        for k in range(x, y): 
            labels = KMeans(n_clusters=k).fit(songs_drp_three_quantile).labels_
            sil_score.append(silhouette_score(songs_drp_three_quantile, labels))
        plt.title(f'Silhouette score evolution from {x} cluster to {y} clusters')
        sns.lineplot(x=range(x, y), y=sil_score)
        sns.scatterplot(x=range(x, y), y=sil_score)
        plt.show()
    if x>y:
        for k in range(y, x): 
            labels = KMeans(n_clusters=k).fit(songs_drp_three_maxAbs).labels_
            sil_score.append(silhouette_score(songs_drp_three_maxAbs, labels))
        plt.title(f'Silhouette score evolution from {y} cluster to {x} clusters')
        sns.lineplot(y=range(y, x), x=sil_score)
        sns.scatterplot(y=range(y, x), x=sil_score)
        plt.show()
```
